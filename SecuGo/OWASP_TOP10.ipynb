```python
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
```

```python
driver=webdriver.Chrome('/home/violeta/Downloads/BIG_DATA/chromedriver')
```

```python
driver.implicitly_wait(3)
```

```python
url='https://www.owasp.org/index.php/Top_10-2017_Top_10'
```

```python
driver.get(url)
```

```python
soup=BeautifulSoup(driver.page_source,'lxml')
```

```python
soup
```

```python
top10=list()
```

```python
for e in soup.select('.mw-content-ltr'):
    top10=e.text
```

```python
top10
```

```python
listSoup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
for e in listSoup.select('div.mw-content-ltr a'):
    top10.append(e.text)
```
