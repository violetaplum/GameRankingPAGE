```python
from selenium import webdriver
import webbrowser
import os
from bs4 import BeautifulSoup
driver = webdriver.Chrome('C:\\Users\\mjkim\\PythonCode\\chromedriver')
driver.implicitly_wait(3)
```

```python
appleUrl='http://www.gevolution.co.kr/rank/ios.asp'
```

```python
driver.get(appleUrl)
```

```python
soup = BeautifulSoup(driver.page_source,'lxml')
```

```python
game_names=soup.select('.rank1')
```

```python
game_names
```

```python
game_names_list=list()
```

```python
for e in soup.select('.rank1'):
    game_names_list.append(e.text)
```

```python
game_names_list
```

```python
removedN = list()
```

```python
for e in game_names_list:
    removedN.append(e.replace("\n",""))
```

```python
removedN
```

```python
len(removedN)
```
